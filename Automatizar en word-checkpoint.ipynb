```python
df = pd.read_excel('Alumnos.xlsx')
```

```python
for indice, fila in df.iterrows():
    contenido = {
        'nombre_alumno': fila["Nombre del Alumno"],
        'nota_mat': fila["Mat"],
        'nota_fis': fila["Fis"],
        'nota_qui': fila["Qui"]
    }

    contenido.update(constantes)
    doc.render(contenido)
    doc.save(f"Notas_de_{fila['Nombre del Alumno']}.docx")
    print(contenido)
```
